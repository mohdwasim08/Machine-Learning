## 1. Scatterplot(Numerical-Numerical)
**Defination:**
A scatter plot is a graph used to show the relationship b/w two numerical variables

**Uses**
- Find corr b/w variables
- Detect patterns or trends
- Identify outliers
- Understand relationships in datasets


```python
tips.head()
```

```python
sns.scatterplot(
    x=tips['total_bill'], 
    y=tips['tip'], 
    hue= tips['sex'], # Color grouping 
    style=tips['smoker'],# Marker style
    size=tips['size'], # Marker size
    palette='deep'
    )
```

**Interpretation**
- Higher bill --> generally higher tip
- Male/Female shown by colour
- Smoker/Non Smoker shown by marker style 
- Groupsize shown by poitn size

## 5. HeatMap(Categorical-Categorical)

**Def:**
- A heatmap is a data visualization technique that uses colors to represent value in matrix or table.

**Uses:**
- Identify patterns in data.
- Fibd correlations b/w variables.
- Analyze frequency tabes(crosstab).
- detect high and low values quickly.

**Interpretation:**
- +1 -> strong Positive correlation.
- -1 -> Strong negative correlation.
- 0 -> No correlation.

```python
titanic.head(3)
```

```python
sns.heatmap(
    pd.crosstab # To calculate the freq (count) b/w two cat variables.
            (titanic['Pclass'],titanic['Survived'])
            )
```

**Observation**
- Pclass = 3, survived=0-->higher.
- Pclass = 2, survived=0 --> low
- Pclass = 1, survived=0 --> very low.

It means 1st class Passanger survive.

- Darker Color = Lower Frequency.
- Lighter Color = Higher Frequency.


```python
(titanic.groupby('Pclass')['Survived'].mean()*100) # survival pecentage.
```

## 6.ClusterMap(Categorical-Categorical)

**Def:**
- A clustermap is an advanced heatmap that groups similar rows and columns using hierarchical clustering.

**Uses:**
- Find similar patterns in data.
- Automatically group related categories.
- Visualize clusters with a dendrogram.

**Note:*
- A clustermap os a heatmap with hierarchical clustring.

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
(pd.crosstab(titanic['Parch'],titanic['Survived'],margins=True,normalize='index')*100)
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived'],margins=True,normalize='index')*100)
```
